```python
# 1 заадание , сравнивание платформ
```

```python
df['CTR'] = df['clicks'] / df['views'] 
df['Share_Rate'] = df['shares'] / df['views'] 
df['Engagement_Rate'] = (df['likes'] + df['shares'] + df['comments']) / df['views'] 
```

```python
df
```

```python
#Анализ по платформам
```

```python
platform_metrics = round(df.groupby('platform')[['CTR', 'Share_Rate', 'Engagement_Rate']].mean(),3).reset_index()
```

```python
platform_metrics
```

```python
#столбчатая диаграмма 
bars = plt.bar(platform_metrics['platform'], platform_metrics['Engagement_Rate'], 
               color=['#c44e52','#4c72b0'],  
               edgecolor='black',  
               width=0.6)  

#добавляем значения поверх столбцов
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height + 0.0004,
             f'{height:.3f}',  
             ha='center', va='bottom', fontsize=10)

#подписи
plt.title('Engagement Rate по платформам', pad=20, fontsize=14)
plt.xlabel('Платформа', labelpad=10, fontsize=12)
plt.ylim(0.17, 0.185)  # Ограничиваем ось Y для лучшего отображения

plt.yticks([])
```

```python
# Анализ по типу контент
```

```python
post_mean = round(df.groupby('post_type')['Engagement_Rate'].mean().sort_values(ascending=False).reset_index(), 3)
```

```python
post_mean
```

```python


# Создаем график
plt.figure(figsize=(8, 5))  # Размер графика (ширина, высота)

# Столбчатая диаграмма с настройками
bars = plt.bar(post_mean['post_type'], post_mean['Engagement_Rate'], 
               color=['#c44e52', '#4c72b0','#4c72b0','#4c72b0'],  # Разные цвета
               edgecolor='black',  # Границы столбцов
               width=0.6)  # Ширина столбцов

# Добавляем значения поверх столбцов
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height + 0.0004,
             f'{height:.3f}',  
             ha='center', va='bottom', fontsize=10)

# Настройки внешнего вида
plt.title('Engagement Rate по типам контента', pad=20, fontsize=14)
plt.xlabel('Тип контента', labelpad=10, fontsize=12)
plt.ylim(0.17, 0.18)  # Ограничиваем ось Y для лучшего отображения

# Убираем лишние элементы
plt.grid(axis='y', alpha=0.3)  # Только горизонтальная сетка
plt.tight_layout()  # Автоматическая настройка отступов
plt.yticks([])
```

```python
#Лучший день недели
```

```python
day_mean = round(df.groupby('weekday')['Engagement_Rate'].mean().sort_values(ascending=False).reset_index(), 3)
```

```python

#что бы упорядочно было 
weekday_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
day_mean['weekday'] = pd.Categorical(day_mean['weekday'], categories=weekday_order)
day_mean = day_mean.sort_values('weekday')

plt.figure(figsize=(10, 6))

#столб диаграмма
bars = plt.bar(
    day_mean['weekday'], 
    day_mean['Engagement_Rate'],
    color=['#4c72b0', '#55a868','#4c72b0','#4c72b0','#4c72b0','#4c72b0', '#55a868'],
    edgecolor='black',
    linewidth=1,
    width=0.7,  
    alpha=0.8
)

#значения на столбцы
for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width()/2.,
        height + 0.0004,
        f'{height:.3f}',
        ha='center',
        va='bottom',
        fontsize=10
    )

#подписи
plt.title('Средний Engagement Rate по дням недели', fontsize=14, pad=20)
plt.xlabel('День недели', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.4)

plt.ylim(0.17, 0.185)
plt.yticks([])


```
